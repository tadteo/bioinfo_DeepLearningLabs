## Introduction to PyTorch

### Tensors and Autograd

PyTorch uses **tensors** as its fundamental data structure. Tensors are similar to NumPy arrays but can run on GPUs and support automatic differentiation (autograd).

**Key concepts:**
- **Tensor**: Multi-dimensional array (like NumPy ndarray)
- **Device**: Tensors can live on CPU or GPU (`tensor.to(device)`)
- **Autograd**: PyTorch automatically tracks operations on tensors for gradient computation
- **requires_grad**: When True, PyTorch tracks all operations on the tensor for backpropagation

### Neural Networks in PyTorch

In PyTorch, you define neural networks by:
1. Creating a class that inherits from `nn.Module`
2. Defining layers in `__init__()`
3. Implementing the forward pass in `forward()`

Example:
```python
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.layer1 = nn.Linear(784, 128)
        self.activation = nn.ReLU()
    
    def forward(self, x):
        x = self.layer1(x)
        x = self.activation(x)
        return x
```

### Training Loop Structure

PyTorch requires you to write the training loop explicitly:

```python
for epoch in range(num_epochs):
    for batch_data, batch_labels in dataloader:
        # 1. Zero gradients
        optimizer.zero_grad()
        
        # 2. Forward pass
        outputs = model(batch_data)
        
        # 3. Compute loss
        loss = criterion(outputs, batch_labels)
        
        # 4. Backward pass (compute gradients)
        loss.backward()
        
        # 5. Update parameters
        optimizer.step()
```

### DataLoader

`DataLoader` handles batching, shuffling, and parallel data loading:
```python
dataset = TensorDataset(features, labels)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```
