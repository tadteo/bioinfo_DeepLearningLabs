# PyTorch Basics for Neural Networks

## Key PyTorch Components:

### 1. nn.Module
In PyTorch, you define models by creating a class that inherits from `nn.Module`. This is more flexible than using `nn.Sequential`.

**Example:**
```python
class MyModel(nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.layer1 = nn.Linear(input_size, hidden_size)
        self.activation = nn.Tanh()
    
    def forward(self, x):
        x = self.layer1(x)
        x = self.activation(x)
        return x
```

### 2. Loss Functions in PyTorch

PyTorch provides several loss functions in `torch.nn`:

- **`nn.MSELoss()`** - Mean Squared Error: Calculates the average of squared differences between predictions and targets. Best for regression problems.

- **`nn.L1Loss()`** - Mean Absolute Error (MAE): Calculates the average of absolute differences. More robust to outliers than MSE.

- **`nn.CrossEntropyLoss()`** - Combines LogSoftmax and NLLLoss. Used for multi-class classification. Expects class indices as targets (NOT one-hot encoded).

### 3. Training Loop
Unlike the high-level `model.fit()` API, PyTorch requires you to write the training loop manually:

```python
for epoch in range(num_epochs):
    for batch_x, batch_y in dataloader:
        optimizer.zero_grad()      # Clear previous gradients
        outputs = model(batch_x)   # Forward pass
        loss = criterion(outputs, batch_y)  # Compute loss
        loss.backward()            # Backward pass (compute gradients)
        optimizer.step()           # Update weights
```

### 4. Regularization in PyTorch

- **Dropout**: Use `nn.Dropout(p=0.2)` as a layer in your model
- **L2 Regularization (Weight Decay)**: Add `weight_decay` parameter to optimizer
  - Example: `optim.SGD(model.parameters(), lr=0.01, weight_decay=0.001)`

````python
# Noise injection function
def stripe_noise(data):
    """
    Adds stripe noise to MNIST images.
    Randomly selects either a horizontal or vertical stripe and sets it to white.
    """
    corrupted = data.clone()
    for ex in range(len(data)):
        ov = np.random.choice([0, 1])
        idx = np.random.choice(np.arange(28))
        if ov == 1:
            corrupted[ex, 0, idx, :] = 1.0  # Horizontal stripe (set to white)
        else:
            corrupted[ex, 0, :, idx] = 1.0  # Vertical stripe (set to white)
    return corrupted

# Visualize noise effect
sample_data = next(iter(test_loader))[0][:5]
corrupted_sample = stripe_noise(sample_data)

plt.figure(figsize=(10, 4))
for i in range(5):
    plt.subplot(2, 5, i + 1)
    plt.imshow(sample_data[i].squeeze(), cmap='gray')
    plt.title('Original')
    plt.axis('off')
    
    plt.subplot(2, 5, i + 6)
    plt.imshow(corrupted_sample[i].squeeze(), cmap='gray')
    plt.title('Corrupted')
    plt.axis('off')

plt.tight_layout()
plt.show()
````
